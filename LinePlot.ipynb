```python
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=y_1, hue="sex", size=10, style="sex", palette="Accent", markers=["o", "^"], dashes=False,
             legend="full")
plt.grid()
plt.title("Python")
plt.xlabel("Bill Length")
plt.ylabel("Flipper Length")

plt.show()
```
