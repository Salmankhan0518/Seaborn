## Multiple Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
fg = sns.FacetGrid(var, col="sex", hue="day")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

### Scatter Plot, differentiate day wise and catagorized by sex  

```python
fg = sns.FacetGrid(var, col="day", hue="sex")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

### Bar Plot, differentiate day wise and catagorized by sex 

```python
fg = sns.FacetGrid(var, col="day", hue="sex")
fg.map(plt.bar, "total_bill", "tip").add_legend()
plt.show()
```

### Bar Plot, differentiate sex wise and catagorized by day

```python
fg = sns.FacetGrid(var, col="sex", hue="day", palette="Accent")
fg.map(plt.bar, "total_bill", "tip", edgecolor="g").add_legend()
plt.show()
```
