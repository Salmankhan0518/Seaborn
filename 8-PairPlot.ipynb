## Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = pd.read_csv("tips.csv").head(20)
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var, vars=["total_bill", "tip"], hue="sex", hue_order=['Male', 'Female'], palette="spring")
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=['Male', 'Female'], x_vars=['total_bill', 'tip'])
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=['Male', 'Female'], y_vars=['total_bill', 'tip'])
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=['Male', 'Female'], markers=["*", "^"])
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=['Male', 'Female'], kind="kde", diag_kind="hist")
plt.show()
```
