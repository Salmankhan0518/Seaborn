## Violin Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="time")
plt.show()
```

### how many people pay bill in different days

```python
sns.violinplot(x="day", y="total_bill", data=var, linewidth=2, palette="Dark2", saturation=0.5)
plt.show()
```

### How many people have lunch or dinner in specific time

```python
sns.violinplot(x="time", y="total_bill", data=var, order=['Dinner', 'Lunch'], inner="quart")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="sex", split=True)
plt.show()
```

### Single Violine plot

```python
sns.violinplot(x=var["total_bill"])
plt.show()
```
