## Cat Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = pd.read_csv("tips.csv").head(20)
var
```

### In cat plot with the help of kind parameter we can drive different plot by using one function

```python
sns.catplot(x="tip", y="size", data=var, hue="sex", height=6, kind="bar")
plt.show()
```
