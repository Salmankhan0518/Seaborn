## Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var
```

```python
sns.set(style="whitegrid")

sns.boxplot(x='day', y='total_bill', data=var, color="r", hue="time", order=['Thur', 'Fri', 'Sat', 'Sun'], showmeans= True, 
            meanprops={"marker": "+", "markeredgecolor": "b"}, linewidth=3, palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x='total_bill', y='day',  data=var, color="r", order=['Thur', 'Fri', 'Sat', 'Sun'], showmeans= True, 
            meanprops={"marker": "+", "markeredgecolor": "b"}, linewidth=3, palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(y=var["total_bill"])
plt.show()
```
