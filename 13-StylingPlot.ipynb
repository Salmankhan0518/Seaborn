## Styling Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = pd.read_csv("tips.csv").head(20)
var
```

```python
sns.set_style("white")
plt.figure(figsize=(3,12))
sns.barplot(x="day", y="total_bill", data=var)

sns.despine()
plt.show()
```

```python
sns.set_context("poster", font_scale=0.2)
sns.barplot(x="day", y="total_bill", data=var)

plt.show()
```
