## Count Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Import csv file, Large data

```python
var = pd.read_csv("tips.csv")
var
```

### Implementation

```python
sns.countplot(x="sex", data=var, hue="smoker", color="g", saturation=0.1)
plt.title('Count Plot')
plt.grid()
plt.show()
```

### Difference btw Count plot and Barplot

```python
sns.barplot(x="sex", y="size", data=var)
plt.title('Bar Plot')
plt.show()
```
