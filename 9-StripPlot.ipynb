## Strip Plot

## Defination:
### A Strip plot is very simple to understand. It is basically a scatter plot that differentiates different categories. Strip plots are considered a good alternative to a box plot or a violin plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = pd.read_csv("tips.csv")
var
```

```python
m={"Male": "o", "Female": "<"}
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="Dark2", linewidth=1, edgecolor="r", jitter=0.2,
              size=5, marker="<", alpha=0.5)
plt.show()
```

```python
sns.stripplot(x=var["total_bill"])
plt.show()
```
