## Bar Plot introduction in Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Import Large Data

```python
var = pd.read_csv("penguins.csv")
var
```

## Without hue comperison basis on sex

```python
sns.barplot(x="island", y="bill_length_mm", data=var)
plt.grid()
plt.show()
```

### Seprate basis on Sex bar chart 

```python
order_1 = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order_1, hue_order=['FEMALE', 'MALE'])
plt.grid()
plt.show()
```

```python
order_1 = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x="bill_depth_mm", y="bill_length_mm", data=var, orient='h')
plt.title('Bar Plot Orient Horizontal')
plt.show()
```
