## Heatmap Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var, cmap="gist_heat", annot=True)
plt.show()
```

```python
data = pd.read_csv("anagrams.csv").head(10)
x = data.drop(columns="attnr")
x
```

```python
sns.heatmap(x, vmin=0, vmax=12, cmap="gist_heat")
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1

y={"fontsize": 20, "color": "r"}

v = sns.heatmap(var1, vmin=0, vmax=12, cmap="PuOr", annot=True, annot_kws=y, linewidths=2, linecolor="y", cbar=False, 
            xticklabels=False, yticklabels=False)
v.set(xlabel = "X Axis", ylabel = "Y Axis")
sns.set(font_scale=5)

plt.show()
```

```python
ar = np.array([['a1', 'a2', 'a3', 'a4', 'a5'], ['b1', 'b2', 'b3', 'b4', 'b5']])
sns.heatmap(var1, vmin=0, vmax=12, cmap="PuOr", annot=ar, fmt='s')
plt.show()
```
